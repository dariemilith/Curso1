## Conceptos básicos en Estadística

- Conceptos basicos:poblacion, muestra, variable, datos
- Tipos de datos: cualitativos y cuantitativos
- Medidas de dispersión: rango, varianza y desviación estándar.


**Población:**

La población es el conjunto completo de elementos o individuos que tienen una característica común y que son objeto de estudio. Por ejemplo, si estamos interesados en la edad de todos los empleados de una empresa, la población sería todos los empleados de esa empresa.

**Muestra:**

La muestra es un subconjunto **representativo** de la población. En muchos casos, es impráctico o imposible estudiar toda la población, ya sea, por tiempo o porque casi siempre es costoso, por lo que se toma una muestra para hacer inferencias sobre la población.

Para que la muestra sea representativa, los elementos seleccionados deben reflejar de manera precisa la variabilidad y características clave de la población completa. Para lograrlo, la selección de la muestra debe realizarse de manera aleatoria y equitativa, asegurando que cada elemento tenga la misma oportunidad de ser incluido. Una muestra representativa permite generalizar los resultados obtenidos a partir de la muestra al conjunto más amplio de la población. La muestra debe reflejar fielmente a la población.


**Variable:**

Una variable es una característica que se mide a cada elemento de la población o muestra. Puede ser cualitativa (categórica) o cuantitativa (numérica). Por ejemplo, a la población de empleados, les medimos la edad, esta sería la variable de tipo cuantitativo.

**Datos:**

Los datos son la información recopilada de las variables. Pueden ser representados en forma de números, palabras o símbolos, y son la materia prima para realizar análisis estadísticos. En nuestro ejemplo, la variable edad, cuando se recolecta de la población o muestra, se convierte en datos.

Para comenzar a ver ejemplos en python y R, introduciremos el concepto de dataframe. Un DataFrame es una estructura de datos bidimensional utilizada en programación para organizar datos en filas y columnas, similar a una tabla u hoja de cálculo. Es comúnmente empleado en análisis de datos usando librerías  como pandas en Python.

Supongamos que tenemos un dataframe con las variables edad y genero de 30 empleados de la empresa A. En este caso, los 30 empleados de la empresa A, serían nuestra población.

```python
#Creación del dataframe
import pandas as pd

# Creación de listas con los datos para las variables edades y genero
lista_edades = [28, 35, 42, 31, 50, 26, 38, 45, 29, 34, 40, 48, 33, 30, 36, 41, 27, 39, 43, 32]
lista_genero = ['M', 'M', 'F', 'F', 'F', 'M', 'F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'M', 'F', 'M', 'M', 'F', 'F']

# Crear un diccionario con las listas
data = {
    'edades': lista_edades,
    'genero': lista_genero
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Ver tipo de datos de las variables
df.head()
```

Seleccionaremos una muestra de tamaño 10 de estos datos utilizando la función **sample()** de la biblioteca pandas. Esta función elige una muestra aleatoria y representativa, teniendo en cuenta las variables del DataFrame, asegurando así la conservación de la distribución de las variables de la población. Al asignar una semilla al parámetro **random_state**, garantizamos que la misma muestra seleccionada sea reproducible en diferentes ejecuciones, ya que la muestra es aleatoria.

```python
muestra = df.sample(n= 10, random_state = 21)
muestra
```


## Tipos de Datos: Cualitativos y Cuantitativos

**Cualitativos:**

Este tipo de datos, describen cualidades o características y no tienen un valor numérico asociado. Por ejemplo, nombres, profesión, categorías.

**Cuantitativos:**

Este tipo de datos, representa cantidades y se expresa en términos numéricos.Estos pueden ser discretos (números enteros) o continuos (números decimales). Por ejemplo, datos discretos son: número de autos o números de hijos, y datos continuos: altura en metros o edad de los empleados.

Los tipos de datos de un dataframe lo podemos ver usando la función **info()** de pandas. En el ejemplo de los empleados, tenemos 2 variables, edad es de tipo cuantitativo (int64) y genero es de tipo cualitativo (object).

```python
df.info()
```

Para un correcto análisis e interpretación de los datos, siempre debemos revisar que nuestro dataframe contenga el correcto tipo de datos de cada variable.
